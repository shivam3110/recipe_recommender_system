### Recipe recommender

```python
import numpy as np 
import pandas as pd
```

```python
file_path = r'D:\shivam\UKW_work\data\food_com_GeniusKitchen\RAW_recipes.csv'
df = pd.read_csv(file_path)
```

```python
df=df[['ingredients','name','id','minutes','tags','steps']]
df.head()
```

```python
file_path = r'D:\shivam\UKW_work\data\food_com_GeniusKitchen\RAW_interactions.csv'
df_inter = pd.read_csv(file_path)
```

```python
df_inter.head()
```

```python
df_inter=df_inter.groupby('recipe_id').agg({'rating':'mean'}).reset_index()
```

```python
df_merged= pd.merge(df[['ingredients','name','id','minutes','tags','steps']],df_inter,how='left',left_on='id',right_on='recipe_id')
```

```python
df_merged
```

```python
df_merged.isnull().sum()
```

```python
df_merged.dropna(subset='name',inplace=True)
```

```python
df_merged.isnull().sum()
```

```python
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def parse_ingredients(ingredients_str):
    return ast.literal_eval(ingredients_str)

df_merged['ingredients_list'] = df_merged['ingredients'].apply(parse_ingredients)
```

```python
def join_ingredients(ingredients):
    return [ingredient.replace(' ', '') for ingredient in ingredients]

df_merged['ingredients_list'] = df_merged['ingredients_list'].apply(join_ingredients)
```

```python
df_merged['ingredients_str'] = df_merged['ingredients_list'].apply(lambda x: ' '.join(x))
```

```python
cv = CountVectorizer(max_features=20000,stop_words='english')
vectors= cv.fit_transform(df_merged['ingredients_str'])
```

```python
def suggest_recipes(user_ingredients):
  input_ingredients_str = ' '.join([ingredient.replace(' ', '') for ingredient in user_ingredients])
  input_vector = cv.transform([input_ingredients_str])
  # Compute cosine similarity between the input vector and recipe vectors
  similarity_scores = cosine_similarity(input_vector, vectors).flatten()
  df_merged['similarity'] = similarity_scores
  top_recipes = df_merged[df_merged.rating>3].nlargest(10, 'similarity')[['name', 'minutes', 'tags', 'steps','rating','similarity']].sort_values(by=['rating'],ascending=False)
  return top_recipes
```

```python
user_ingredients =['winter squash', 'butter', 'olive oil']
suggested_recipes = suggest_recipes(user_ingredients)
(suggested_recipes)
```
